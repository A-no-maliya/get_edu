```python
import json
import matplotlib.pyplot as plt
import statistics
from matplotlib.ticker import MaxNLocator


file  = open('try is not try.json', encoding="UTF-8" )
data = json.load(file)


class Student:
    def __init__(self, name, age, height):
        self.name = name
        self.age = age
        self.height = height

    def __repr__(self):
        return f'{{"Name":"{self.name}", "Age":"{self.age}", "Height":"{self.height}"}}'
    
    
data_dict = dict()
ages, heights, names = list(), list(), list()

for i in data.items():
    tmp = Student(i[1]["Name"], int(i[1]["Age"]), int(i[1]["Height"]))
    names.append(i[1]["Name"])
    ages.append(int(i[1]["Age"]))
    heights.append(int(i[1]["Height"]))
    data_dict[i[0]] = tmp


ax = plt.subplot()


ax.xaxis.set_major_locator(MaxNLocator(12))
ax.xaxis.set_tick_params(labelsize=6.5)


ax.plot(names, ages, label="Ages")
ax.set_ylabel("Ages and Height")
ax.plot(names, heights, label ="Height")
ax.plot(names,[sum(ages) / len(ages)]*13, label= 'average ages')
ax.plot(names,[sum(heights) / len(heights)]*13, label='average heights')
ax.plot(names, [statistics.pstdev(heights)]*13)
ax.plot(names, [statistics.pstdev(ages)]*13)


plt.show()

print(data_dict)

sorted_dict = dict(sorted(data_dict.items(), key=lambda x: x[1].age))

print(sorted_dict)

names, ages, heights = list(), list(), list()

for i in sorted_dict.items():
    names.append(i[1].name)
    ages.append(int(i[1].age))
    heights.append(int(i[1].height))
image.png

bx = plt.subplot()

bx.xaxis.set_major_locator(MaxNLocator(12))
bx.xaxis.set_tick_params(labelsize=6.5)

bx.plot(names, heights, label='Height')
bx.plot(names, ages, label='Weight')
bx.plot(names, ages, label="Ages")
bx.set_ylabel("Ages and Height")
bx.plot(names, heights, label ="Height")

plt.show()


with open('sorted_class.json', 'w' , encoding="UTF-8") as new_file:
    sorted_dict = str(sorted_dict).replace("'", '"')
    new_file.write(sorted_dict)
```
